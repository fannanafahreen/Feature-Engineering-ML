# Ordinal Encoder

```python
df['income'].unique()
```

```python
income_order=['low','medium','high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories=[income_order])
enco.fit(df[['income']])
```

```python
encode_data = enco.transform(df[['income']])
encode_data
```

```python
final=pd.DataFrame(encode_data,columns=['income'])
final
```

```python
df['income']=final
df
```
