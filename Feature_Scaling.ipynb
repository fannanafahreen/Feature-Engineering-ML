```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_set = pd.read_csv('E:/dataset/mock_dataset.csv')
data_set.head(10)
```

```python
x= data_set[['Age','Salary']]
x
```

# min max scaler (Normalization )

```python
x_temp = x.copy()
x_temp['Age'] = (x_temp['Age']-x_temp['Age'].min()) / (x_temp['Age'].max()-x_temp['Age'].min())
x_temp
```

```python
sns.distplot(x_temp['Age'])
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
x_after_min_max_scaler = MMS.fit_transform(x)
df=pd.DataFrame(x_after_min_max_scaler,columns =["Age","Salary"])
df
```

# standardization

```python
x_temp = x.copy()
x_temp['Age'] = (x_temp['Age']- x_temp['Age'].mean())/ x_temp['Age'].std()
x_temp
```

```python
sns.distplot(x_temp['Age'])
```

```python
from sklearn.preprocessing import StandardScaler as SC
std= SC()
x_after_standardisation = std.fit_transform(x)
df=pd.DataFrame(x_after_standardisation)
df
```
